# 📌 1⃣ Import Libraries

```python
# Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 📌 2⃣ Read Dataset

```python
Data=pd.read_csv("IPL 2022.csv")
Data.head(5)
```

# 📌 3⃣ Matches Won by Each Team (Matplotlib)

```python
# Count matches won by each team
wins = Data['match_winner'].value_counts()

plt.figure(figsize=(10,5))
plt.bar(wins.index, wins.values, color='orange')
plt.xticks(rotation=45)
plt.title("Number of Matches Won in IPL 2022")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.tight_layout()
plt.show()
```

# 📌 4⃣ Matches Won by Defending or Chasing (Seaborn)

```python
Data = pd.read_csv("IPL 2022.csv")
Data['won_by'] = Data['won_by'].str.strip().str.title().map({'Wickets':'Chasing','Runs':'Defending'}).fillna('Other')

ax = sns.countplot(x='won_by', data=Data, order=Data['won_by'].value_counts().index)

# Add count numbers on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Matches Won by Defending or Chasing (IPL 2022)")
plt.ylabel("Number of Matches")
plt.xlabel("Match Result Type")
plt.tight_layout()
plt.show()
```

# 🎯 5⃣ Best Bowling Players (Seaborn)

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y='best_bowling',
    data=Data,
    order=Data['best_bowling'].value_counts().index[:10]
)
plt.title("Top 10 Best Bowlers in IPL 2022")
plt.xlabel("Count")
plt.ylabel("Player")
plt.show()
```

# 🏅 6⃣ Player of the Match (Matplotlib)

```python
# Count top 10 players with most 'Player of the Match' awards
player_counts = Data['player_of_the_match'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.barh(player_counts.index, player_counts.values, color='lightblue')
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Awards")
plt.ylabel("Players")
plt.show()
```

# 💥 7⃣ Top Scorers (Seaborn + NumPy)

```python
top_scores = Data.groupby("top_scorer")["highscore"].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_scores.values, y=top_scores.index)
plt.title("Top 10 Scorers in IPL 2022")
plt.xlabel("High Score")
plt.ylabel("Player")
plt.show()
```
